```python
# Eliminar aquella vivienda que no tiene precio, area ni districto
%pip install pandas
import pandas as pd
df=pd.read_csv('barcelona_viviendas_dataset.csv', sep=',')
df.info()

```

```python
df=df[df['area'].notnull()]
df=df[df['room_num'].notnull()]
df.info()
df.head()
```

Sacar estadística para cada barrio,  que tendrá los siguientes campos:

- **barrio**:nombre de barrio
- **total_area_for_sale**: area de vivienda en venta
- **num_sale**: número de viviendas en venta
- **mean_price**:precio medio de las viviendas
- **median_price**: la mediana de los precio de las viviendas
- **max_price**: precio máximo
- **min_price**: precio mínimo
- **mean_price_m2**:precio medio de las viviendas por m2
- **median_price_m2**:la mediana de los precio por m2 de las viviendas
- **max_price_m2**:precio máximo m2
- **min_price_m2**:precio mínimo m2
- **mean_m2_price_with_elevator**: precio medio de viviendas con ascensor
- **mean_m2_price_with_terrace**: precio medio de vieviendas con terraza
- **mean_m2_price_with_attic**: precio medio de viviendas con atico
- **mean_m2_price_with_parking**: precio medio de viviendas con parking
- **mean_m2_price_with_furnished**: precio medio de vivienda con muebles
- **frecuent_room_num**: número de habitaciones más comunes
- **frecuent_area**: area de viviendas más comunes

```python
import numpy as np
distric_data = {}

for index, row in df.iterrows():
    uid = row['uid']
    barrio_name = row['district']
    area = row['area']
    room_num = row['room_num']
    description = row['description']
    price = row['price_euro']
    price_m2 = price/area
    furnished = row['furnished']
    detail = row['features_detail']
    ascensor = row['has_elevator']
    parking = row['has_parking']
    if barrio_name not in distric_data:
        distric_data[barrio_name] = {
            'uid': [uid],
            'area': [area],
            'barrio': [barrio_name],
            'room_num': [room_num],
            'description': [description],
            'price': [price],
            'price_m2': [price_m2],
            'furnished': [furnished],
            'features_detail': [detail],
            'ascensor': [ascensor],
            'parking': [parking]

        }
    else:
        distric_data[barrio_name]['uid'].append(uid)
        distric_data[barrio_name]['area'].append(area)
        distric_data[barrio_name]['barrio'].append(barrio_name)
        distric_data[barrio_name]['room_num'].append(room_num)
        distric_data[barrio_name]['description'].append(description)
        distric_data[barrio_name]['price'].append(price)
        distric_data[barrio_name]['price_m2'].append(price_m2)
        distric_data[barrio_name]['furnished'].append(furnished)
        distric_data[barrio_name]['features_detail'].append(detail)
        distric_data[barrio_name]['ascensor'].append(ascensor)
        distric_data[barrio_name]['parking'].append(parking)

results = []
for barrio_name, values in distric_data.items():

    with_elevator_price = []
    with_attic_price = []
    with_terrace_price = []
    with_parking_price = []
    with_furnish_price = []
    # parking
    for idx, parking in enumerate(values['parking']):
        if parking == True:
            with_parking_price.append(values['price_m2'][idx])
    # ascensor
    for idx, ascensor in enumerate(values['ascensor']):
        if ascensor == True:
            with_elevator_price.append(values['price_m2'][idx])

    for idx, mueble in enumerate(values['furnished']):
        if mueble == True:
            with_furnish_price.append(values['price_m2'][idx])

    for idx, description in enumerate(values['description']):
        # atico
        if 'atico' in description.lower() or 'ático' in description.lower():
            with_attic_price.append(values['price_m2'][idx])
        # terraza
        if 'terraza' in description.lower() and 'sin terraza' not in description.lower():
            with_terrace_price.append(values['price_m2'][idx])

    area_10_scale = []
    for area in values['area']:
        area_10_scale.append(area//10)

    most_frecuent_area = pd.value_counts(area_10_scale).index[0]*10

    results.append({
        'barrio_name': barrio_name,
        'num_sale': len(values['uid']),
        'total_area_for_sale': int(sum(values['area'])),
        'mean_price': int(np.mean(values['price'])),
        'median_price': int(np.median(values['price'])),
        'max_price': max(values['price']),
        'min_price': min(values['price']),
        'mean_price_m2': int(np.mean(values['price_m2'])),
        'median_price_m2': int(np.median(values['price_m2'])),
        'max_price_m2': int(max(values['price_m2'])),
        'min_price_m2': int(min(values['price_m2'])),
        'mean_m2_price_with_elevator': int(np.nan_to_num(np.mean(with_elevator_price))),
        'mean_m2_price_with_terrace': int(np.nan_to_num(np.mean(with_terrace_price))),
        'mean_m2_price_with_attic': int(np.nan_to_num(np.mean(with_attic_price))),
        'mean_m2_price_with_parking': int(np.nan_to_num(np.mean(with_parking_price))),
        'mean_m2_price_with_furnished': int(np.nan_to_num(np.mean(with_furnish_price))),
        'frecuent_room_num': int(pd.value_counts(values['room_num']).index[0]),
        'frecuent_area': int(most_frecuent_area)
    })

result_data=pd.DataFrame(results)
```
